```python
import pandas as pd
```

```python
# read csv
df = pd.read_csv('windowed_files.csv')
df
```

```python
def set_val(df, val_percent):
    df = df.copy()
    indices_to_update = []
    print("Determining indices to update...")
    for label in df["label"].unique():
        non_test = df[(~df["split"].str.contains("test")) & (df["label"] == label)]
        print(f"Label: {label}, Count of non-'test' rows: {len(non_test)}")
        sample_indices = non_test.sample(frac=val_percent).index
        indices_to_update.extend(sample_indices)
        print(f"Label: {label}, Selected indices for 'val': {len(sample_indices)}")
    
    # Update the DataFrame in one go
    print("\nUpdating 'split' column in df...")
    df.loc[indices_to_update, "split"] = "val"
    
    # Count and print the number of 'val' rows for each label after the update
    print("\nFinal counts of 'val' rows per label:")
    for label in df["label"].unique():
        val_count = len(df[(df["label"] == label) & (df["split"] == "val")])
        print(f"Label: {label}, Count of 'val': {val_count}")

    return df

# Example usage
set_val(df, 0.2)
```
